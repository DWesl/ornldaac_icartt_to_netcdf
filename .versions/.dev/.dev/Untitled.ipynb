```python
!ls
import pandas as pd
from glob import glob
import os
```

```python
csv = pd.read_csv("VARIABLES.csv")
csv
```

```python
variable_names = [os.path.splitext(os.path.basename(f))[0] for f in glob("../variables/*")]
variable_names
```

```python
matches = []

for i, row in csv.iterrows():
    icartt_name = row.ICARTT_NAME
    if icartt_name in variable_names:
        match = [i for i in variable_names if i==icartt_name][0] + ".json"
        matchlist = row.tolist()
        matchlist[0] = match
        matches.append(matchlist)
```

```python
with open("VARIABLES_match.csv", "w") as f:
    for m in matches:
        m = [str(i) for i in m]
        f.write(",".join(m) + "\n")
```
